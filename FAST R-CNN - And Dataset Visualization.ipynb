# Train and Save Fast R-CNN

```python
import os

cfg = get_cfg()
cfg.merge_from_file(model_zoo.get_config_file("COCO-Detection/faster_rcnn_R_50_FPN_3x.yaml"))

cfg.DATASETS.TRAIN = ("garbage_train",)
cfg.DATASETS.TEST = ("garbage_valid",)

cfg.DATALOADER.NUM_WORKERS = 4

cfg.MODEL.WEIGHTS = model_zoo.get_checkpoint_url("COCO-Detection/faster_rcnn_R_50_FPN_3x.yaml")

cfg.SOLVER.IMS_PER_BATCH = 6
cfg.SOLVER.BASE_LR = 0.01
cfg.SOLVER.MAX_ITER = 1000
cfg.SOLVER.STEPS = [20000, 25000]         # Learning rate decay
cfg.SOLVER.WARMUP_ITERS = 100

cfg.MODEL.ROI_HEADS.BATCH_SIZE_PER_IMAGE = 64
cfg.MODEL.ROI_HEADS.NUM_CLASSES = 5

cfg.OUTPUT_DIR = "./output"

os.makedirs(cfg.OUTPUT_DIR, exist_ok=True)

cfg.MODEL.WEIGHTS = model_zoo.get_checkpoint_url("COCO-Detection/faster_rcnn_R_50_FPN_3x.yaml")
print("Using model weights:", cfg.MODEL.WEIGHTS)

trainer = DefaultTrainer(cfg)
trainer.resume_or_load(resume=False)
trainer.train()
```

# Test Model

```python
# Function to compute IoU
def compute_iou(boxA, boxB):
    """
    Compute Intersection over Union (IoU) between two bounding boxes.
    Boxes are in the format [x_min, y_min, width, height].
    """
    xA = max(boxA[0], boxB[0])
    yA = max(boxA[1], boxB[1])
    xB = min(boxA[0] + boxA[2], boxB[0] + boxB[2])
    yB = min(boxA[1] + boxA[3], boxB[1] + boxB[3])

    interArea = max(0, xB - xA) * max(0, yB - yA)
    boxAArea = boxA[2] * boxA[3]
    boxBArea = boxB[2] * boxB[3]

    iou = interArea / float(boxAArea + boxBArea - interArea)
    return iou


# Main script
import json
import numpy as np
from collections import defaultdict
from pycocotools.coco import COCO
import pandas as pd

# Load predictions and ground truth annotations
predictions_path = "./output/coco_instances_results.json"  # Path to saved predictions
ground_truth_path = "Data Annotate Garbage Bags.v4i.coco/test/_annotations.coco.json"  # Path to ground truth

with open(predictions_path, "r") as f:
    predictions = json.load(f)
ground_truth = COCO(ground_truth_path)

# Initialize a dictionary to store results
results_per_class = defaultdict(lambda: {"TP": 0, "FP": 0, "FN": 0})

# Define IoU threshold for a "true positive"
iou_threshold = 0.5

# Process each image in the dataset
for image_id in ground_truth.getImgIds():
    # Get ground truth annotations for this image
    gt_ann_ids = ground_truth.getAnnIds(imgIds=image_id)
    gt_annotations = ground_truth.loadAnns(gt_ann_ids)

    # Get predictions for this image
    image_predictions = [pred for pred in predictions if pred["image_id"] == image_id]

    # Match predictions to ground truth
    matched_gt = set()  # Track matched ground truth annotations
    for pred in image_predictions:
        pred_bbox = pred["bbox"]
        pred_class = pred["category_id"]
        pred_score = pred["score"]

        max_iou = 0
        matched_gt_id = None

        # Compare prediction with each ground truth annotation
        for gt in gt_annotations:
            if gt["category_id"] != pred_class:  # Skip if classes don't match
                continue

            gt_bbox = gt["bbox"]

            # Compute IoU
            iou = compute_iou(pred_bbox, gt_bbox)
            if iou > max_iou and iou >= iou_threshold and gt["id"] not in matched_gt:
                max_iou = iou
                matched_gt_id = gt["id"]

        if matched_gt_id:
            # True Positive: Match found
            results_per_class[pred_class]["TP"] += 1
            matched_gt.add(matched_gt_id)
        else:
            # False Positive: No match found
            results_per_class[pred_class]["FP"] += 1

    # Count False Negatives (ground truth annotations not matched to any prediction)
    for gt in gt_annotations:
        if gt["id"] not in matched_gt:
            gt_class = gt["category_id"]
            results_per_class[gt_class]["FN"] += 1

# Calculate precision and recall for each class
categories = ground_truth.loadCats(ground_truth.getCatIds())
for cat in categories:
    class_id = cat["id"]
    metrics = results_per_class[class_id]
    tp = metrics["TP"]
    fp = metrics["FP"]
    fn = metrics["FN"]
    precision = tp / (tp + fp) if (tp + fp) > 0 else 0
    recall = tp / (tp + fn) if (tp + fn) > 0 else 0
    metrics["Precision"] = precision
    metrics["Recall"] = recall

# Convert results to a pandas DataFrame for better visualization
data = []
for cat in categories:
    class_id = cat["id"]
    class_name = cat["name"]
    metrics = results_per_class[class_id]
    data.append([
        class_name,
        metrics["TP"],
        metrics["FP"],
        metrics["FN"],
        metrics["Precision"],
        metrics["Recall"]
    ])

df_metrics = pd.DataFrame(data, columns=["Class", "TP", "FP", "FN", "Precision", "Recall"])
print("\nPer-Class Metrics:")
print(df_metrics)

# Optional: Add overall averages
overall_precision = df_metrics["Precision"].mean()
overall_recall = df_metrics["Recall"].mean()
df_metrics.loc["Overall"] = ["Overall", df_metrics["TP"].sum(), df_metrics["FP"].sum(), df_metrics["FN"].sum(), overall_precision, overall_recall]

print("\nPer-Class Metrics with Overall Summary:")
print(df_metrics)
```
